```python
import re
def ifpyt(search1):
    regex="^if\((\w+)+[>,<,=,!]+\w+\):\s+\w+\s+(elif?\(\w+[<,>,=,!]+\w+\):\s+\w+)?\s+else:\s+\w+"
    ifinp=re.search(regex,search1)
    if(ifinp):
        print("You entered If for python statement")
    else:
        print("You entered wrong statement")
def ifjava(search2):
    regex="^if[(][\w]+[<,>,=,!]+\w+\)\{\s+\w+\s+\}\s+else\{\s+\w+\s+\}"
    ifjinp=re.search(regex,search2)
    if(ifjinp):
        print("You entered If for java statement" )
    else:
        print("You entered wrong statement")
def forpyt(search3):
    regex="^for[ ][\w+][ ]in[ ][\w]+[:][A-Za-z\s,.?!:]+"
    forinp=re.search(regex,search3)
    if(forinp):
        print("You entered For for python statement")
    else:
        print("You entered wrong statement")
def forjava(search4):
    regex="^for[(][\w]+[ ][\w]+[<,>,=,!]+[\d]+[;][ ][\w]+[<,>,=,!]+.*[\w]+[;][ ][\w]+[+]+[)][{].*[\s]+.*[\w]+.*[\s]+[}]$"
    forjinp=re.search(regex,search4)
    if(forjinp):
        print("You entered For for java statement")
    else:
        print("You entered wrong statement")
def whilepyt(search5):
    regex="^while[ ][\w+][ ][<,>,=,!]+[ ][\d]+[:][\s]+.*[\s]+"
    whileinp=re.search(regex,search5)
    if(whileinp):
        print("You entered While for python statement")
    else:
        print("You entered wrong statement")
def whilejava(search6):
    regex="^while[ ][(][\w+][ ][<,>,=,!]+.*[\w]+[)][ ][{][\s].*[\s]+[\w]+[+]+[;].*[\s].*[}]$"
    whilejinp=re.search(regex,search6)
    if(whilejinp):
        print("You entered While for java statement")
    else:
        print("You entered wrong statement")
def dowhilepyt(search7):
    regex="^while True[:].*[\s]+.*[\w]+[(][\w]+[)].*[\s]+[\w]+[ ][=].*[\w]+.*[\s]+.*[\s]+break$"
    dowhileinp=re.search(regex,search7)
    if(dowhileinp):
        print("You entered Do While for python statement")
    else:
        print("You entered wrong statement")
def dowhilejava(search8):
    regex="^do.*[\s]+[{].*[\s]+[\w]+.*[(][\w]+[)][;].*[\s]+.*[}].*(while[(][\w]+[)][;])$"
    dowhilejinp=re.search(regex,search8)
    if(dowhilejinp):
        print("You entered Do While for java statement")
    else:
        print("You entered wrong statement")
def switchpyt(search9):
    regex="^switch\(\w+\)\s+\{\s+|((case)\s+\w+:\s+\w+\s+\"\w+\";\s+)+default:\s+\w+\s+\"\w+\";\s+\};"
    switchinp=re.search(regex,search9)
    if(switchinp):
        print("You entered Switch for python statement")
    else:
        print("You entered wrong statement")
def switchjava(search10):
    regex="^switch\(\w+\)\s+\{\s+|(case\s+\w+:\s+\w+;\s+break;\s+)+default:\s+(\s+\w+;\s+)+break;\s\}"
    switchjinp=re.search(regex,search10)
    if(switchjinp):
        print("You entered Switch for java statement")
    else:
        print("You entered wrong statement")
def identifier(search11):
    regex="[A-Za-z\s+]+"
    ideninp=re.search(regex,search11)
    if(ideninp):
        print("You entered Identifier statement")
    else:
        print("You entered wrong statement")
def datatype(search12):
    regex="(class\{|\s+|double|\s+|int|\s+|float|\s+|boolean|\s+|})"
    dtinp=re.search(regex,search12)
    if(dtinp):
        print("You entered Data types statement")
    else:
        print("You entered wrong statement")
def main():
    inp=input("Enter Code")
    code=inp
    val=re.findall(r'[A-Za-z\s+\d+]+',inp)
    value=re.findall(r'(class\{|\s+|double|\s+|int|\s+|float|\s+|boolean|\s+|})',inp)
    if(code[0:2]=="if"):
        ifpyt(inp)
    elif(code[0:3]=="for"):
        forpyt(inp)
    elif(code[0:5]=="while"):
        whilepyt(inp)
    elif(code[0:2]=="do"):
        dowhilepyt(inp)
    elif(code[0:5]=="switch"):
        switchpyt(inp)
    elif(val):
        identifier(inp)
    elif(value):
        datatype(inp)
    elif(code[0:3]=="if("):
        ifjava(inp)
    else:
        print("Enter Right code")
        main()
main()
```
